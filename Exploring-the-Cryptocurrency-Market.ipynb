```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
plt.style.use('fivethirtyeight')
```

```python
data = pd.read_csv("/Users/akhilesh/Downloads/projects/coinmarketcap.csv")
```

```python
data.head()
```

```python
market_cap_raw = data[['id','market_cap_usd']]
market_cap_raw.count()
```

```python
TOP_CAP_TITLE = 'Top 10 market capitalization'

cap10 = data.head(10).set_index(data.id[:10])
cap10 = cap10.assign(market_cap_percentage = lambda x: (x.market_cap_usd / data.market_cap_usd.sum()) * 100)
```

```python
cap10
```

```python
colors = ['orange', 'green', 'orange', 'cyan', 'cyan', 'blue', 'silver', 'orange', 'red', 'green']

bar = cap10.plot.bar(x='id',y='market_cap_percentage',color=colors,title='TOP_CAP_TITLE')
bar.set_ylabel('% of total cap')
bar.set_xlabel('')
```

```python
volatility = data[['id','percent_change_24h', 'percent_change_7d']]
volatility = volatility.set_index('id').dropna()
volatility
```

```python
volatility1d= volatility.sort_values('percent_change_24h',ascending = True)
volatility1d.head()
```

```python
def top10_subplot(volatility_series, title):
    fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 6))
    ax = volatility_series[:10].plot.bar(color = 'darkred', ax = axes[0])
    fig.suptitle(title)
    ax.set_ylabel('% change')
    ax.set_xlabel('')
    ax = volatility_series[-10:].plot.bar(color = 'darkblue', ax = axes[1])
    ax.set_xlabel('')
    return fig, ax

day_title = "24 hours top losers and winners"
 
fig, ax = top10_subplot(volatility1d.percent_change_24h, day_title)
```

```python
volatility7d = volatility.sort_values('percent_change_7d', ascending = True)
volatility7d 
```

```python
wtitle = 'weekly top winners and loosers'

fig,ax = top10_subplot(volatility7d.percent_change_7d,wtitle)
```

```python
def capcount(info):
    return data.query(info).count().id

largecap = capcount('market_cap_usd > 300000000')

midcap = capcount('market_cap_usd < 300000000 and market_cap_usd > 50000000')

smallcap = capcount('market_cap_usd < 50000000')

values = [largecap,midcap,smallcap]

plt.bar(range(len(values)), values, tick_label = ['largecap','midcap','smallcap'],color=['red','green','black'])
```
